```python
from geopy.geocoders import Nominatim
```

```python
def geocoding(address):
    while True:
        geolocoder = Nominatim(user_agent = 'South Korea', timeout=None)
        geo = geolocoder.geocode(address)
        if geo == None:
            print(address)
            a = "1"
            return a
            break
        crd = [str(geo.latitude), str(geo.longitude)]#float로 하면 메모리 많이 잡아먹으니까 str으로 저장하고 사용할 때 float로 변경
        return crd
        break
```

```python
geo_x_y = []
for i in wide['도로명']:
    a = geocoding(i)
    if a == "1":#지오코드로 위도경도가 없다면
        geo_x_y.append(geo_x_y[-1])#바로 이전에 추가한 위도경도를 추가한다.
    else:
        geo_x_y.append(a)
```
