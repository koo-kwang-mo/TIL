## 앞에 붙어있는 정보들이 인덱스였다 -> 열로 바꿔주자, 바꾼 후 pop2로 다시 저장

```python
pop2 = pop.reset_index()
```

```python
pop2 = pop2.rename(columns = {'level_0' : 'Rank;County;State;Land Area', 'level_1' : '2020 census', 'level_2' : '2021 Estimates', 'level_3' : '2020 census','level_4' : '2021 Estimates'}, inplace = False)
#열 이름 붙이기
```

```python
pop2
```

## 1번 열에 4가지 값이 섞여있다 분해해보자

```python
pop3 = pd.DataFrame(pop2['Rank;County;State;Land Area'].str.split(';').tolist(), columns = ['Rank','Country','State','LandArea'])
```

## 각 행의 값이 ;으로 분리돼 있어 ;를 기준으로 잘라서 이름을 다시 붙여보지만 5개의 값이 있다고 한다.

```python
a = pd.DataFrame(pop2['Rank;County;State;Land Area'].str.split(';').tolist())
a
```

## 왜 5개라고 하는가 했더니 중간에 값이 5개나 나오는 행이 있다. 어떻게 하지? 

## 값을 살펴보니 5개가 있는 경우에는 4번째 값을 지우고 5번째 값이 4번째에 가야된다.
### => 5번쨰에 none이면 4번째 값을 복사해서 넣고 난 다음 4번째행을 모두 없애자.

```python
a.fillna(a[3])
```

## 각행의 3인덱스가 아니라 처음걸로 다 채워진다..

```python
for i in range(100):
  if a[4][i] == None:
    a[4][i] = a[3][i]
```

```python
a
```

## 다행이 행의 길이가 짧고 값을 안다
# 굿

```python
a1 = a.drop(3, axis = 1)
a2 = a1.rename(columns = {'0' :'Rank','1':'Country','2':'State','4':'LandArea'}, inplace = False)
a2
```
