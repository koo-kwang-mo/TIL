## 바탕이 투명한 이미지를 적용시키자

```python
#4채널이 됐을 떄 기존 3채널 이미지와 붙이는 함수

def overlay(image,x, y, w, h, overlay_image):#대상 이미지, x, y, 좌표, width, height, 덮어씌울 이미지(4채널)s
    alpha = overlay_image[:,:,3]# BGRA, 마지막은 알파값
    mask_image = alpha /255 # 0~255 -> 0~1 값으로 바뀜, 1:불투명 0: 투명
        
        
    """
    255,255 -> 1, 1
    255, 0 -> 1, 0
    1-mask_image
    0, 0
    0, 1
     (1 - mask_image) 부분의 의미 : 원래 이미지의 색도 어느정도 인용함
     불투명 : 오버레이 이미지 가져옴
     투명한 영역 : 대상이미지 가져옴       
    """
        
        
                
        
    for c in range(0,3): #BGR 순서대로 처리할 예정, A무시
        image[y-h:y+h, x-w:x+w, c] = (overlay_image[:,:,c]*mask_image) + (image[y-h:y+h,x-w:x+w,c] * (1 - mask_image))
        #0~2채널에만 적용
```
